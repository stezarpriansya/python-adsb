## How value of b affect success probability

### 1.1 With b>0

```python
#Define function b>0

def reglog_pos(x):
    return math.exp(0.01 + 0.8*x) / (1 + math.exp(0.01 + 0.8*x))
```

```python
reglog_pos(1)
```

```python
func_pos = np.vectorize(reglog_pos)
```

### Example 1

```python
x1 = np.random.randn(1000)
```

```python
plt.hist(x1)
plt.show()
```

```python
pi_pos = func_pos(x1)
```

```python
plt.scatter(x1,pi_pos)
plt.show()
```

peningkatan x juga peluang semakin meningkat tapi bukan linear --> sigmoid

```python
#Define function b>0

def reglog_zero(x):
    return math.exp(0.01 + 0*x) / (1 + math.exp(0.01 + 0*x))
```

```python
func_pos = np.vectorize(reglog_zero)
x1 = np.random.randn(1000)
pi_pos = func_pos(x1)
```

```python
plt.scatter(x1,pi_pos)
plt.show()
```

```python
p_success = func_neg(1)
print(p_success)

p_failure = 1 - func_neg(1)
print(p_failure)

odd = p_success/p_failure
print('odd:',odd)

print(1/odd)
```

Parameter tuning merupakan metode yang digunakan untuk meningkatkan nilai akurasi setelah dilakukan feature selection terhadap data. parameter tuning digunakan untuk memiliki hasil dari parameter model yang dapat meningkatkan akurasi.

```python
import numpy as np
from sklearn.grid_search import GridSearchCV
from sklearn import datasets, svm
import matplotlib.pyplot as plt
```

```python
digits = datasets.load_digits()
# Create dataset 1
datatrain = data_baru_train.drop('target',axis=1)[:7000]
train_target = dftrain['target'][:7000]

# Create dataset 2
datatest = data_baru_test.drop('target',axis=1)[7000:]
test_target = y2[7000:]
```

Sebelum memilih kombinasi dari parameter yang menghasilkan model paling akurasi, kita perlu membedakan nilai yang diolah

```python
# Create a classifier object with the classifier and parameter candidates
clf = GridSearchCV(estimator=svm.SVC(), param_grid=parameter_candidates, n_jobs=-1)

# Train the classifier on data1's feature and target data
clf.fit(datatrain, train_target) 
```

```python
# View the accuracy score
print('Best score for data1:', clf.best_score_) 
```

```python
# Apply the classifier trained using data1 to data2, and view the accuracy score
clf.score(datatest, test_target)
```

# Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
a = np.random.rand(10,1)
b = np.random.rand(10,1)
c = np.random.rand(10,1) 
```

## Basic

```python
plt.plot(a,b, color = 'turquoise') #(x,y,color)
plt.plot(a,c, color = (1.0, 0.2, 0.3))
plt.xlabel('X')
plt.title('Random')
plt.show()
```

```python
plt.axes([0.05,0.05,0.425,0.9]) #[position x, position y, size x, size y]
plt.plot(a,b,'red')
plt.xlabel('X')
plt.title('Random 1')
plt.axes([0.6,0.05,0.425,0.9])
plt.plot(a,c,'blue')
plt.xlabel('X')
plt.title('Random 2')
plt.show()
```

## Subplot

```python
plt.subplot(2,1,1) #baris, kolom, posisi
plt.plot(a,b,'red')
plt.xlabel('X')
plt.title('Random 1')
plt.subplot(2,1,2)
plt.plot(a,c,'blue')
plt.xlabel('X')
plt.title('Random 2')

#untuk merapikan
plt.tight_layout()
plt.show()
```

```python
plt.subplot(2,2,2) #baris, kolom, posisi
plt.plot(a,b,'red')
plt.xlabel('X')
plt.title('Random 1')
plt.subplot(2,2,3)
plt.plot(a,c,'blue')
plt.xlabel('X')
plt.title('Random 2')

#untuk merapikan
plt.tight_layout()
plt.show()
```

## Axis
Seperti zoom di kamera

```python
plt.plot(a,b,'red')
plt.xlabel('X')
plt.title('Random')
# plt.xlim((0.2,0.4))
# plt.ylim((0.2,0.6))
plt.axis((0.2,0.4,0.2,0.6))
plt.show()
```

## Legend

```python
plt.plot(a,b,'red', label = 'red')
plt.plot(a,c,'blue', label = 'blue')
plt.xlabel('X')
plt.title('Random')
plt.axis('tight')
# Legend
plt.legend(loc = 'lower right')
plt.show()
```

## Annotation

```python
plt.plot(a,b,'red', label = 'red')
plt.plot(a,c,'blue', label = 'blue')
plt.xlabel('X')
plt.title('Random')
plt.axis('tight')

plt.legend(loc = 'upper right')
# Annotation with arrow
plt.annotate('red', xy = (0.4,0.1),xytext=(0.3,0.01),arrowprops={'color':'red'}) 
#xy itu buat pointing kemana, xytext itu tempat buntut
plt.annotate('blue', xy = (0.5,0.7),xytext=(0.55,0.75),arrowprops={'color':'blue'})
plt.show()
```

## Style

```python
plt.style.use('grayscale')
plt.plot(a,b,label = 'red')
plt.plot(a,c,label = 'blue')
plt.xlabel('X')
plt.title('Random')
plt.axis('tight')

plt.legend(loc = 'upper right')
# Annotation with arrow
plt.annotate('black', xy = (0.4,0.17),xytext=(0.3,0.01),arrowprops={'color':'black'}) 
#xy itu buat pointing kemana, xytext itu tempat buntut
plt.annotate('grey', xy = (0.5,0.7),xytext=(0.55,0.75),arrowprops={'color':'grey'})
plt.show()
```

## DATA AUTO

```python
auto = pd.read_csv("Data Python 3/auto/imports-85.data.txt", index_col = None)
auto.shape
```

```python
auto.isnull().sum()
```

```python
auto.head()
```

```python
#handling missing value
for i in [1, 5, 18, 19, 21, 22, 25] :
    auto[auto.columns[i]] = np.where(auto[auto.columns[i]] == "?", np.nan, auto[auto.columns[i]])

auto.head()
#convert to float per column
column_float = [1, 9, 10, 11, 12, 13, 16, 18, 19, 20, 21, 22, 23, 24, 25]
for i in column_float :
    auto[auto.columns[i]] = auto[auto.columns[i]].astype(float)
    auto[auto.columns[i]] = auto[auto.columns[i]].fillna(auto[auto.columns[i]].mean())

auto.head()
```

```python
auto.isnull().sum()
```

```python
auto.dtypes
# 1. symboling:                -3, -2, -1, 0, 1, 2, 3.
#   2. normalized-losses:        continuous from 65 to 256.
#   3. make:                     alfa-romero, audi, bmw, chevrolet, dodge, honda,
#                                isuzu, jaguar, mazda, mercedes-benz, mercury,
#                                mitsubishi, nissan, peugot, plymouth, porsche,
#                                renault, saab, subaru, toyota, volkswagen, volvo
#   4. fuel-type:                diesel, gas.
#   5. aspiration:               std, turbo.
#   6. num-of-doors:             four, two.
#   7. body-style:               hardtop, wagon, sedan, hatchback, convertible.
#   8. drive-wheels:             4wd, fwd, rwd.
#   9. engine-location:          front, rear.
#  10. wheel-base:               continuous from 86.6 120.9.
#  11. length:                   continuous from 141.1 to 208.1.
#  12. width:                    continuous from 60.3 to 72.3.
#  13. height:                   continuous from 47.8 to 59.8.
#  14. curb-weight:              continuous from 1488 to 4066.
#  15. engine-type:              dohc, dohcv, l, ohc, ohcf, ohcv, rotor.
#  16. num-of-cylinders:         eight, five, four, six, three, twelve, two.
#  17. engine-size:              continuous from 61 to 326.
#  18. fuel-system:              1bbl, 2bbl, 4bbl, idi, mfi, mpfi, spdi, spfi.
#  19. bore:                     continuous from 2.54 to 3.94.
#  20. stroke:                   continuous from 2.07 to 4.17.
#  21. compression-ratio:        continuous from 7 to 23.
#  22. horsepower:               continuous from 48 to 288.
#  23. peak-rpm:                 continuous from 4150 to 6600.
#  24. city-mpg:                 continuous from 13 to 49.
#  25. highway-mpg:              continuous from 16 to 54.
#  26. price:                    continuous from 5118 to 45400.
```

```python
# convert door from categorical to numerical
def doors_num(banyak):
    if banyak == 'four':
        return 4.0
    elif banyak == 'two':
        return 2.0
    else :
        return np.nan
```

```python
# doors_num('four')
auto['num-of-doors'] = list(map(doors_num, auto['num-of-doors']))
auto.head()
```

```python
sns.countplot(y='make', data=auto, orient='h')
```

```python
sns.countplot(x='fuel-type', data=auto)
```

```python
sns.countplot(x='num-of-doors', data=auto)
```

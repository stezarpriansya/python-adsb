```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from IPython.display import Image
```

```python
%%markdown
## With b>0
```

```python
# define function with b>0

def reglog_pos(x):
    return math.exp(0.01+0.8*x) / (1+ math.exp(0.01 + 0.8*x))
```

```python
reglog_pos(1)
```

```python
func_pos = np.vectorize(reglog_pos)
```

```python
x1 = np.random.randn(1000)
```

```python
plt.hist(x1)
plt.show()
```

```python
pi_pos = func_pos(x1)
```

```python
plt.scatter(x1, pi_pos)
plt.show()
```

```python
x2 = np.random.uniform(-10,10,1000)
```

```python
plt.hist(x2)
plt.show()
```

```python
plt.scatter(x2, func_pos(x2))
plt.show()
```

```python
%%markdown
## With b<0
```

```python
# define function with b<0

def reglog_neg(x):
    return math.exp(0.01+(-0.8*x)) / (1+ math.exp(0.01 + (-0.8*x)))
```

```python
func_neg = np.vectorize(reglog_neg)
```

```python
x3 = np.random.randn(1000)
```

```python
pi_neg = func_neg(x3)
```

```python
plt.scatter(x3, pi_neg)
plt.show()
```

```python
x4 = np.random.uniform(-10,10,1000)
```

```python
plt.scatter(x4, func_neg(x4))
plt.show()
```

```python
%%markdown
## With b = 0
```

```python
# define function with b=0

def reglog_zero(x):
    return math.exp(0.01+(0*x)) / (1+ math.exp(0.01 + (0*x)))
```

```python
func_zero = np.vectorize(reglog_zero)
x5 = np.random.randn(1000)
```

```python
plt.scatter(x5, func_zero(x5))
plt.show()
```

```python
x6 = np.random.uniform(-10,10,1000)
```

```python
plt.scatter(x6, func_zero(x6))
plt.show()
```

```python
%%markdown
With b>0, p(x) increase with the increase of x <br/>
With b>0, p(x) decrease with the increase of x <br/>
With b=0, p(x) constant for any number of x <br/>
```

```python
%%markdown
# ODD
ODD adalah peluang sukses dibandingkan peluang gagal dari sebuah X<br/>
ODD = Phi(x) / 1 - Phi(x) = E^(alfa + beta*x)
```

```python
%%markdown
## Example 1
Y "success" = 1 <br/>
Y "failure" = 0
```

```python
# peluang 'success' when x = 1

p_success = func_pos(1)
p_success

# peluang telat bayar 0.69 jadi dia telat

# peluang 'failure' when x = 1

p_failure = 1- func_pos(1)
p_failure

# peluang dia bayar tepat waktu 0.307

# ODD when x = 1

ODD = p_success / p_failure
ODD
```

```python
%%markdown
## Example 2
```

```python
# peluang 'success' when x = 1

p_success = func_neg(1)
p_success

# peluang 'failure' when x = 1

p_failure = 1- func_neg(1)
p_failure


# ODD when x = 1

ODD = p_success / p_failure
ODD
```

```python
%%markdown
## Excercise 1
```

```python
#tenor pembayaran
# peluang 'success' when x = 2

p_success = func_pos(2)
print(p_success)

# peluang telat bayar 0.83 jadi dia telat

# peluang 'failure' when x = 2

p_failure = 1- func_pos(2)
print(p_failure)

# peluang dia bayar tepat waktu 0.16

# ODD when x = 2

ODD = p_success / p_failure
print(ODD) #peluang bayar jika tenor peminjaman 2 bulan adalah 5 kali dibanding tepat bayar
```

```python
%%markdown
## Excercise 2
```

```python
#turn over karyawan
# peluang 'success' when x = 2

p_success = func_neg(2)
print(p_success)

# peluang telat bayar 0.16 jadi dia tidak telat

# peluang 'failure' when x = 2

p_failure = 1- func_neg(2)
print(p_failure)

# peluang dia bayar tepat waktu 0.83

# ODD when x = 2

ODD = p_success / p_failure
print(ODD) #kalo nilai ODD kecil, interpretasi dapat dibaca dengan dibalik
print(1/ODD) # peluang seseorang tetap kerja 4.9 kali lebih besar dibandingkan dia resign setelah 2 tahun kerja
```

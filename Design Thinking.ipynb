```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline
```

```python
penumpang = pd.read_csv("data_bus_sekolah_2017/data_penumpang_bus_2017.csv")
```

```python
bulan = bulan = ['januari', 'februari', 'maret', 'april', 'mei', 'juni', 'juli', 'agustus', 'september', 'oktober', 'november', 'desember']
for bul in bulan:
    penumpang[bul] = penumpang[bul].astype('float64')
```

```python
penumpang1 = penumpang[['daerah_operasi', 'januari', 'februari', 'maret', 'april', 'mei', 'juni', 'juli', 'agustus', 'september', 'oktober', 'november', 'desember']]
```

```python
penumpang1.dtypes
```

```python
bus = pd.read_csv("data_bus_sekolah_2017/jumlah_bus_2017.csv")
```

```python
for bul in bulan:
    bus[bul] = bus[bul].astype('float64')
```

```python
bus1 = bus[['daerah_operasi', 'januari', 'februari', 'maret', 'april', 'mei', 'juni', 'juli', 'agustus', 'september', 'oktober', 'november', 'desember']]
```

```python
jumlah = 0
for bul in bulan:
    jumlah = jumlah + bus1[bul]
bus1['rata2_tahun'] = jumlah/12
```

```python
bus1
```

```python
jumlah = 0
for bul in bulan:
    jumlah = jumlah + penumpang1[bul]
penumpang1['rata2_tahun'] = jumlah/12
```

```python
gabung = penumpang[['daerah_operasi']]
```

```python
gabung['kepadatan_tahun'] = round(penumpang1['rata2_tahun'] / bus1['rata2_tahun'])
```

```python
gabung = gabung.set_index(['daerah_operasi'])
```

```python
gabung[['kepadatan_tahun']].sort_values(by=['kepadatan_tahun'], ascending= False).plot.bar(title ='Kepadatan Bus per Zona/Rute', figsize = (15,10), color = ('green'))
plt.hlines(y=173, xmin=0, xmax=50)
plt.show()
```

```python
penumpang_tok = penumpang1[['daerah_operasi', 'rata2_tahun']]
penumpang_tok = penumpang_tok.set_index('daerah_operasi')
```

```python
penumpang_tok[['rata2_tahun']].sort_values(by=['rata2_tahun'], ascending= False).plot.bar(title ='Rata2 penumpang Bus per Zona/Rute', figsize = (15,10), color = ('red'))
plt.hlines(y=7000, xmin=0, xmax=50)
plt.show()
```

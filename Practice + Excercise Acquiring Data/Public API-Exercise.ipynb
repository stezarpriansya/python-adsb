1. Calculate the increasement(or decreasement) from opening price to closing price each day.
2. Determine the highest closing price & lowest volume among those 5 days

```python
# Hint: find the stock symbol for Astra International
### Fill the code ###
data = {
    "apikey":API_KEY,
    "function": 'TIME_SERIES_DAILY',
    "symbol": 'AUTO',
    'datatype': 'json'
}
```

```python
headers = {'Content-Type': 'application/json'}
```

```python
response = requests.get(API_BASE_URL, params = data, headers = headers)
```

```python
### Fill the code ###
# Convert to json
response = response.json()
```

```python
print(pprint.pprint(response))
```

```python
### Fill the code ###
# Get the daily response only

daily = response['Time Series (Daily)']
key = list(daily.keys())
```

```python
### Fill the code ###
# Get the five days data
days = []
for i in range(0,5) :
    days.append({key[i] : daily.get(key[i])})
```

```python
### Fill the code ###
# calculate the difference
for day in days:
    this_key = list(day.keys())[0]
    print('Tanggal '+ this_key)
    open_price = float(day[this_key]['1. open'])
    close_price = float(day[this_key]['4. close'])
    diff = open_price-close_price
    print('difference: {0}'.format(diff))
```

# TASK 2

```python
closing = {}
volume = {}
```

```python
### Fill the code ###
for day in days:
    # assign to closing object
    this_key = list(day.keys())[0]
    closing[this_key] = float(day[this_key]['4. close'])
```

```python
# Iterate over object
highest_closing_value = 0.0;
highest_closing_attr = None;
for attr, value in closing.items():
    # compare highest and assign
    if value > highest_closing_value:
        highest_closing_value = value
        highest_closing_attr = attr
        
print('Nilai closing tertinggi ada pada tanggal '+highest_closing_attr+ ' dengan nilai '+str(highest_closing_value))
```

```python
### Fill the code ###
for day in days:
    # assign to volume object
    this_key = list(day.keys())[0]
    volume[this_key]= float(day[this_key]['5. volume']) #ambil seluruh nilai volume last 5 days
```

```python
# Iterate over object
lowest_volume_value = 0.0;
lowest_volume_attr = None;
for attr, value in volume.items():
    # compare lowest and assign
    if value > lowest_volume_value:
        lowest_volume_value = value
        lowest_volume_attr = attr
        
print('Nilai closing tertinggi ada pada tanggal '+lowest_volume_attr+ ' dengan nilai '+str(lowest_volume_value))
```

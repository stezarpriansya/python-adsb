```python
import pandas as pd 
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df1= pd.read_csv('bus.csv', index_col=0)
```

```python
df1.head()
```

```python
df1[['rataratabus']].sort_values(by=['rataratabus'], ascending= False).plot.bar(title ='Rata rata bus', figsize = (12,5), color = ('purple'))
```

```python
df1[['rataratap']].sort_values(by=['rataratap'], ascending= False).plot.bar(title ='Rata rata penumpang', figsize = (12,5), color = ('pink'))
```

```python
df1['kepadatan'] = round(df1['rataratap']/df1['rataratabus'])
```

```python
df1.head()
```

```python
df1[['kepadatan']].sort_values(by=['kepadatan'], ascending= False).plot.bar(title ='kepadatan bus sekolah', figsize = (12,5), color = ('lightblue'))
x= df1['kepadatan'].mean()
print(x)
plt.hlines(y=154, xmin=0, xmax=50)
plt.show()
```

```python
# Created by Stezar
import pandas as pd
import numpy as np
```

```python
pets1 = pd.read_excel("Pets1.xlsx")
pets2 = pd.read_excel("Pets2.xlsx")
owners = pd.read_csv("Owners.csv")
```

```python
# pets1 = pets1.set_index("PetID")
# pets2 = pets2.set_index("PetID")
# owners = owners.set_index("OwnerID")
```

```python
pets = pd.concat([pets1,pets2])
```

```python
print('Pets Shape', pets.shape)
print('Owners Shape', owners.shape)
```

```python
print('Pets NA')
print(pets.isnull().sum())
```

```python
print('Owners NA')
print(owners.isnull().sum())
```

```python
owners.info()
len(owners.OwnerID.unique())
```

```python
pets.info()
```

```python
pets.Gender = np.where(pets['Kind'] == "Dog", pets.Gender.fillna('male'), pets.Gender)
pets.Gender = np.where(pets['Kind'] == "Cat", pets.Gender.fillna('female'), pets.Gender)
pets.Gender = np.where(pets['Kind'] == "Parrot", pets.Gender.fillna('male'), pets.Gender)
# pets
```

```python
mean_age = pets.groupby(['Kind', 'Gender'])['Age'].mean()
mean_age
```

```python
pets.Age = np.where((pets['Kind'] == "Dog") & (pets['Gender'] == "male"), pets.Age.fillna(value=mean_age['Dog']['male']), pets.Age)
pets.Age = np.where((pets['Kind'] == "Dog") & (pets['Gender'] == "female"), pets.Age.fillna(value=mean_age['Dog']['female']), pets.Age)
pets.Age = np.where((pets['Kind'] == "Cat") & (pets['Gender'] == "male"), pets.Age.fillna(value=mean_age['Cat']['male']), pets.Age)
pets.Age = np.where((pets['Kind'] == "Cat") & (pets['Gender'] == "female"), pets.Age.fillna(value=mean_age['Cat']['female']), pets.Age)
pets.Age = np.where((pets['Kind'] == "Parrot") & (pets['Gender'] == "male"), pets.Age.fillna(value=mean_age['Parrot']['male']), pets.Age)
pets.Age = np.where((pets['Kind'] == "Parrot") & (pets['Gender'] == "female"), pets.Age.fillna(value=mean_age['Parrot']['female']), pets.Age)
```

```python
pets
```

```python
pets_owner = pd.merge(owners,pets,how='inner',on='OwnerID')
```

```python
# pets_owner.head(3)
print(pets_owner.isnull().sum())
```

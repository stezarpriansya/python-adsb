### Function
    Function ada dua :
    built-in function 
    User defined function

```python
#User defined function
def label_height(height):
    """this to label height"""
    if height > 150:
        return "tinggi"
    else: 
        return "rendah"
```

```python
print(label_height(170))
```

```python
#map used for iterates all element in list to the func
print(list(map(label_height, Fam)))
print(list(map(label_height, [row[1] for row in Data]))) #datanya dilekularkan dari list di dalam list
```

```python
#Lambda is anonymous func
print(list(map(lambda x : x + 100, [row[1] for row in Data])))
```

### Looping

### Create Data Frame from Dict

```python
code = ['US', 'AUS', 'JPN', 'IN', 'RUS', 'MOR', 'EGY']
ctr = ['United States', 'Australia', 'Japan', 'India', 'Russia', 'Morocco', 'Egypt']
dr = [True, False, False, False, False, True, True]
cpc = [809, 731, 588, 18, 200, 70, 45]

my_dict = {'ctrcode': code, 'country': ctr, 'drivers_right': dr, 'cars_per_cpc': cpc}
```

```python
# create dataframe
import pandas as pd
cars = pd.DataFrame(my_dict)
cars
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.models import model_from_json
import numpy as np
from keras.callbacks import ModelCheckpoint
```

```python
np.random.seed(7)
```

```python
dataset = np.loadtxt('dataset_dl/pima-indians-diabetes/pima-indians-diabetes.csv',delimiter=',')
```

```python
X = dataset[:, 0:8] #fitur
y = dataset[:,8] #target
```

```python
#buat model
model = Sequential()
model.add(Dense(12, input_dim=8, kernel_initializer='uniform', activation='relu'))
model.add(Dense(8, activation='relu', kernel_initializer='uniform'))
model.add(Dense(1, activation='sigmoid', kernel_initializer='uniform'))

#compile model
model.compile(loss='binary_crossentropy',optimizer = 'adam', metrics=['accuracy'])

filepath = 'improvisasi-bobot-epoch{epoch:02d}-akurasi-{val_acc:.2f}.hdf5'
checkpoint = ModelCheckpoint(filepath, monitor="val_acc", verbose=1, save_best_only=True, mode='max')
callbacks_list = [checkpoint]

#fit model
model.fit(X,y, validation_split=0.33, epochs=150, batch_size=10, verbose=0, callbacks=callbacks_list)

#evaluasi model
# scores = model.evaluate(X,y, verbose=0)
# print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
#serialize model ke json
model_json = model.to_json()
with open("model_nn.json", "w") as json_file:
    json_file.write(model_json)

#serialize wwight ke hdf5
model.save_weights("weight_nn.h5")
print("[INFO] Menyimpan model....")
```

## Load Model

```python
from keras.models import model_from_json
```

```python
# load Model
json_file = open('model_nn.json', 'r')
loaded_from_json = json_file.read()
json_file.close()
loaded_model = model_from_json(loaded_from_json)
```

```python
# Load weight
loaded_model.load_weights("improvisasi-bobot-epoch146-akurasi-0.78.hdf5")
print('[INFO] Model & Weight Loaded...')
```

```python
#compile model
loaded_model.compile(loss='binary_crossentropy',optimizer = 'adam', metrics=['accuracy'])

#fit model
loaded_model.fit(X,y, epochs=150, batch_size=10, verbose=0)

#evaluasi model
scores = loaded_model.evaluate(X,y, verbose=0)
print("\n%s: %.2f%%" % (loaded_model.metrics_names[1], scores[1]*100))
```
